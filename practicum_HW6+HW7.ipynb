```python
import pygame
from pygame.draw import *
from random import randint

pygame.init()

FPS = 60
screen = pygame.display.set_mode((0, 0), pygame.FULLSCREEN)
h = screen.get_height()
w = screen.get_width()
WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
YELLOW = (255, 255, 0)
GREEN = (0, 255, 0)
MAGENTA = (255, 0, 255)
CYAN = (0, 255, 255)
BLACK = (0, 0, 0)
COLORS = [RED, BLUE, YELLOW, GREEN, MAGENTA, CYAN]
list_of_balls = []
flag_s = False
time = super_vx = super_vy = super_x = super_y = 0

def create_ball():
    global list_of_balls
    '''создаём новые шары '''
    x = randint(100, 1100)
    y = randint(100, 900)
    r = randint(30, 100)
    vx = randint(-10, 10)
    vy = randint(-10, 10)
    color = COLORS[randint(0, 5)]
    circle(screen, color, (x, y), r)
    list_of_balls.append([x, y, r, vx, vy, color])

pygame.display.update()
clock = pygame.time.Clock()
flag = flag_2 = False
while not flag:
    clock.tick(FPS)
    count = len(list_of_balls) + flag_s  # это число шаров
    for i in pygame.event.get():
        if i.type == pygame.KEYDOWN:
            flag = True
        if i.type == pygame.QUIT:
            flag = True
        elif i.type == pygame.MOUSEBUTTONDOWN:
            if not flag_2:
                mouse_x, mouse_y = i.pos
                new_balls = []
                for el in list_of_balls:
                    if not (el[0] - mouse_x) ** 2 + (el[1] - mouse_y) ** 2 <= (el[2] ** 2):
                        new_balls.append(el)
                list_of_balls = new_balls
            if flag_s:
                if (time // 10) % 2 == 0:
                    if 0 < (-super_x + mouse_x) < 200 and 0 < (-super_y + mouse_y) < (time % 10 * 20 + 20):
                        flag_s = False
    if not flag_2:
        for ball in list_of_balls:
            ball[0] += ball[3]
            ball[1] += ball[4]

            circle(screen, ball[5], (ball[0], ball[1]), ball[2])
            if ball[0] >= w or ball[0] <= 0:
                ball[3] *= -1
            if ball[1] >= h or ball[1] <= 0:
                ball[4] *= -1
        if flag_s:
            if (time // 10) % 2 == 0:
                super_surface1 = pygame.transform.scale(super_surface, (200, time % 10 * 20 + 20))
            else:
                super_surface1 = pygame.transform.scale(super_surface, (time % 10 * 20 + 20, 200))
            super_x += super_vx
            super_y += super_vy
            if super_x >= w or super_x <= 0:
                super_vx *= -1
            if super_y >= h or super_y <= 0:
                super_vy *= -1
            screen.blit(super_surface1, (super_x, super_y))

    if time % 60 == 0 and not flag_2: 
        if count == 4 and not flag_s:
            flag_s = True
            super_x = randint(100, 1100)
            super_y = randint(200, 600)
            super_vx = randint(-10, 10)
            super_vy = randint(-10, 10)
            super_surface = pygame.Surface((200, 200), pygame.SRCALPHA)
            super_surface.fill((255, 255, 255, 0))
            polygon(super_surface, WHITE, [[0, 100], [100, 200], [200, 100], [100, 0]])

        else:
            create_ball()

    pygame.display.update()
    screen.fill(BLACK)
    time += 1
    f1 = pygame.font.Font(None, 36)
    f2 = pygame.font.Font(None, 360)

    text1 = f1.render('Time: ' + str(int(time / 6) / 10), 1, (180, 180, 180))
    if count >= 10:
        flag_2 = True
        screen.fill(BLACK)
        text3 = f2.render('LOSE', 1, (255, 30, 40))
        screen.blit(text3, (w / 2 - 500, h / 2 - 300))

    if count > 0:
        text2 = f1.render('Number of balls: ' + str(count), 1,
                          ((255 * count / 10) % 255, (255 * (10 - count) / 10) % 255, 0))
    else:
        text2 = f1.render('Number of balls: ' + str(count), 1, BLUE)
    if not flag_2:
        screen.blit(text1, (10, 50))
        screen.blit(text2, (w - 300, 50))

pygame.quit()
```
